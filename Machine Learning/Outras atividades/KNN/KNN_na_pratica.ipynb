Prever qual linguagem de programção de um novo funcionário baseado nas informações, como departamento, quantos anos de empresa e assim conseguir estimar qual seria essa linguagem de programação

```python
import pandas as pd
```

```python
base = pd.read_excel("/content/funcionarios.xlsx")
```

```python
base.shape
```

```python
base.head()
```

```python
del base['Unnamed: 0'] # Excluindo coluna que não traz nenhuma informação relevante
```

```python
base.head()
```

```python
#separar colunas de entrada e saída
X = base.iloc[:,:-1] #todas linhas e menos ultima coluna
```

```python
X
```

```python
y = base.iloc[:,-1] #todas linhas e somente a ultima coluna
```

```python
y
```

```python
from sklearn.neighbors import KNeighborsClassifier # Biblioteca de Hiperparametros (numero de vizinhos)
```

```python
knn = KNeighborsClassifier(n_neighbors=3) # criar instancia a partir do classificador para 3 vizinhos
```

```python
knn.fit(X,y) # função responsavel pelo treinamento
```

```python
amostra = base.sample() # verificar se esta funcionando
```

```python
amostra
```

```python
amostra.iloc[:,:-1]
```

```python
knn.predict(amostra.iloc[:,:-1]) #Informa qual a linguagem
```

```python
knn.predict([[2,1,0,0,1,0]]) #Utilizando outros refinamentos da tabela
```
